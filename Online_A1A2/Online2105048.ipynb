```python
import numpy as np

x = 65767879797907
y = 765454532435435345

x_digits = [int(digit) for digit in str(x)]

y_digits = [int(digit) for digit in str(y)]

def fft_1d(signal):
    N = len(signal)
    if N <= 1:
        return signal
    
    if N % 2 != 0:
        raise ValueError("Signal size must be a power of 2 for FFT.")

    even = fft_1d(signal[::2])
    odd = fft_1d(signal[1::2])

    T = [np.exp(-2j * np.pi * k / N) * odd[k] for k in range(N // 2)]
    return [even[k] + T[k] for k in range(N // 2)] + [even[k] - T[k] for k in range(N // 2)]


def ifft_1d(freq_signal):
    N = len(freq_signal)
    if N <= 1:
        return freq_signal
    
    if N % 2 != 0:
        raise ValueError("Signal size must be a power of 2 for IFFT.")

    conj_signal = np.conjugate(freq_signal)

    fft_conjugate = fft_1d(conj_signal)

    return [np.conjugate(x) / N for x in fft_conjugate]

digits_A = x_digits[::-1] 
digits_B = y_digits[::-1] 

n = len(digits_A) + len(digits_B) - 1 
N = 1 << (n - 1).bit_length()  

digits_A = np.pad(digits_A, (0, N - len(digits_A)))
digits_B = np.pad(digits_B, (0, N - len(digits_B)))

fft_A = fft_1d(digits_A)
fft_B = fft_1d(digits_B)

fft_product = [fft_A[k] * fft_B[k] for k in range(N)]

res = ifft_1d(fft_product)
res = np.round(np.real(res)).astype(int)  

result = []
c= 0
for val in res:
    total = val + c
    result.append(total % 10)  
    c= total // 10 

while c > 0: 
    result.append(c % 10)
    c //= 10
        
print("".join(str(d) for d in reversed(result))) 

```
